## Quick Exercise (Normalisation)

### 1. Create a statevector that will give a 1/3 probability of measuring the state $|0\rangle$

$Solution:$  

Whatever state you have constructed, first make sure that the sum of squares of the coefficients (or amplitudes) of $|0\rangle$ and $|1\rangle$ add up to $1$.

Next make sure that the square of the amplitude of $|0\rangle$ is $\large\frac{1}{3}$.

The most straightforward state is: $\Large\frac{1}{\sqrt3}\normalsize|0\rangle + \Large\frac{2}{\sqrt3}\normalsize|1\rangle$

### 2. Create a different statevector that will give the same measurement probabilities.

$Solution:$  

Again, the simplest statevector is one where we just switch the sign between the terms:  

$\Large\frac{1}{\sqrt3}\normalsize|0\rangle - \Large\frac{2}{\sqrt3}\normalsize|1\rangle$

Otherwise,

$\Large\frac{(\sqrt2 + i)}{3}\normalsize|0\rangle \pm \Large\frac{(2+\sqrt2 i)}{3}\normalsize|1\rangle$

### 3. Verify that the probability of measuring  $|1\rangle$  for these two states is  2/3 .

$Solution:$  

Again, the (magnitude of coefficient of $|1\rangle$)$^2$ will give us the required probability.

## Quick Exercise (Bloch Sphere)

### Use ```plot_bloch_vector()``` or ```plot_bloch_sphere_spherical()``` to plot a qubit in the states:

````python
from qiskit_textbook.widgets import plot_bloch_vector_spherical
from math import pi,sqrt
````

### 1. $|0\rangle$

$Solution: $  

We need  
$\normalsize\cos(\large\frac{\theta}{2}\normalsize)|0\rangle + i sin(\large\frac{\theta}{2}\normalsize)|1\rangle = |0\rangle$

which implies $\theta$ must be $0$, and since this eliminates the effect of $\phi$, it doesn't matter what $\phi$ is. We let it remain $0$.

````python
coords = [0,0,1]
plot_bloch_vector_spherical(coords)
````

### 2. $|1\rangle$

We need  
$\normalsize\cos(\large\frac{\theta}{2}\normalsize)|0\rangle + i sin(\large\frac{\theta}{2}\normalsize)|1\rangle = |1\rangle$

which implies $\Large\frac{\theta}{2}$ must be $\Large\frac{\pi}{2}$, and $\phi$ must have no effect on our state so it remains $0$.

````python
coords = [pi,0,1]
plot_bloch_vector_spherical(coords)
````

### 3. $\large\frac{1}{\sqrt2}\normalsize(|0\rangle+|1\rangle)$

$\normalsize cos(\large\frac{\theta}{2}\normalsize) = \large\frac{1}{\sqrt2}$ and $\normalsize sin(\large\frac{\theta}{2}\normalsize) = \large\frac{1}{\sqrt2}$

We need $\phi$ to not affect, hence $\phi = 0$

````python
coords = [pi/2,0,1]
plot_bloch_vector_spherical(coords)
````

### 4. $\large\frac{1}{\sqrt2}\normalsize(|0\rangle-i|1\rangle)$

````python
coords = [pi/2,-pi/2,1]
plot_bloch_vector_spherical(coords)
````

### 5. $\large\frac{1}{\sqrt2}\normalsize\begin{bmatrix}i\\1\end{bmatrix}$

````python
coords = [pi/2,-pi/2,1]
plot_bloch_vector_spherical(coords)
````
